```python
import matplotlib.pyplot as plt
import requests
import time


page_link = 'https://covid.cdc.gov/covid-data-tracker/COVIDData/getAjaxData?id=US_MAP_DATA'


def get_info():
    page = requests.get(page_link)
    response = page.json()
    us_map_data = response.get('US_MAP_DATA')

    # for item in us_map_data:
    #     print(item.get('name'))
    #     print(item.get('Seven_day_avg_new_cases_per_100k'))

    return us_map_data

print('*************')
ax = []
ay = []
for state in get_info():
    ax.append(state.get('abbr'))
    ay.append(state.get('Seven_day_avg_new_cases_per_100k'))

plt.bar(ax, ay)
plt.ylabel('Average Daily Cases per 100k in Last 7 Days')
plt.xlabel('State/Territory')
plt.show()

```
